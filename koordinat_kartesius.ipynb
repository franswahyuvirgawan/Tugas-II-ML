```python
import matplotlib.pyplot as plt

x_axis = [7, 5, -8, -5]
y_axis = [5, -4, -5, 3]
labels = ['Kuadran I', 'Kuadran IV', 'Kuadran III', 'Kuadran II']

plt.title('Koordinat Kartesius')
plt.scatter(x_axis, y_axis, color="darkblue", marker='x',  label="Data Points")

plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')

for i, label in enumerate(labels):
    plt.text(x_axis[i], y_axis[i], label, fontsize=12, ha='right')

plt.grid(True)
plt.legend()

plt.show()
```
